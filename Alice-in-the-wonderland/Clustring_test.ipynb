```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/Datamining/df_scaled.csv')
```

```python
data.head()
data.drop('cluster', axis=1, inplace=True)
```

```python
data.sample(2)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3)
km.fit_predict(data)

data['cluster'] = km.labels_
```

```python
import seaborn as sn
sn.scatterplot(data=data , x='x' , y='y' , hue='cluster')
```
